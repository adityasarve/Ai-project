## Exploratory Data Analysis (EDA):

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('cleaned_delivery_data.csv')
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values)
```

 Check for numerical features.

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include=[np.number])
```

```python
# Calculate summary statistics
summary_stats = numerical_cols.describe()
print(summary_stats)
```

Visualize the distributions of numerical features

```python
# List of numerical features
numerical_features = ['courier_id', 'lng', 'lat', 'accept_gps_lng', 'accept_gps_lat', 'delivery_gps_lng', 'delivery_gps_lat', 'ds']

# Create subplots for each numerical feature
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(12, 15))

# Iterate through features and create plots
for i, feature in enumerate(numerical_features):
    row = i // 2  # Calculate row index
    col = i % 2   # Calculate column index
    sns.histplot(df[feature], kde=True, ax=axes[row, col])
    axes[row, col].set_title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```

Exploreing the distribution of categorical features

```python
top_couriers = df['courier_id'].value_counts().head(10)
plt.figure(figsize=(10, 5))
top_couriers.plot(kind='bar')
plt.title('Distribution of Top 10 Couriers')
plt.xlabel('Courier ID')
plt.ylabel('Count')
plt.show()
```

 The distribution of orders across the top 10 couriers is slightly skewed, with a few couriers handling significantly more orders than others.

Univariate Data Analysis

```python
numerical_columns = ['order_id', 'region_id', 'courier_id', 'lng', 'lat', 'aoi_id',
                     'accept_gps_lng', 'accept_gps_lat', 'delivery_gps_lng', 'delivery_gps_lat', 'ds']

# Loop through each numerical column
for column in numerical_columns:
    plt.figure(figsize=(10, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data=df, x=column, bins=30)
    plt.title(f'Histogram of {column}')

    # Kernel Density Plot
    plt.subplot(1, 2, 2)
    sns.kdeplot(data=df, x=column)
    plt.title(f'Kernel Density Plot of {column}')

    plt.show()

    # Summary Statistics
    print(f'Summary Statistics for {column}:\n')
    print(df[column].describe())
    print('\n')
```

The numerical features in the dataset generally exhibit a relatively even distribution across the range of 0 to 1, with some showing slight peaks in specific areas.

Multivariate Data Analysis

 Correlation Matrix and Heatmap

```python
# Select numerical features
numerical_columns = ['order_id', 'region_id', 'courier_id', 'lng', 'lat',
                     'accept_gps_lng', 'accept_gps_lat', 'delivery_gps_lng', 'delivery_gps_lat', 'ds']

# Calculate correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

The correlation matrix reveals that there is no significant correlation between any of the numerical variables in the dataset. This suggests that the variables are independent of each other.

Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA

# Standardize the data (important for PCA)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[numerical_columns])

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 principal components for visualization
principal_components = pca.fit_transform(scaled_data)

# Create a scatter plot of the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: First Two Principal Components')
plt.show()

# Explained variance ratio
print(pca.explained_variance_ratio_)
```

 The PCA plot shows the distribution of data points in the first two principal components. There appears to be some clustering of data points, indicating potential groupings or patterns in the data. However, without further context or analysis, it is difficult to draw definitive conclusions from this plot.

Scatter Plot

The 3D scatter plot of lng, lat, and ds shows a complex pattern of data points in three dimensions. It's difficult to draw definitive conclusions from this visualization alone, but it suggests that there might be some underlying structure or relationships between these variables. Further analysis, such as clustering or dimensionality reduction techniques, may be needed to uncover more insights.
